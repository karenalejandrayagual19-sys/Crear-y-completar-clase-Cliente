```

class Estudiante:
    def __init__(self, nombre: str, apellido: str, cedula: str, email: str = None, semestre: int = None, materias: tuple = None):
        self._nombre = nombre
        self.apellido = apellido
        self.cedula = cedula
        self._semestre = None
        self.semestre = semestre
        self.email = email
        self.materias = materias

    # Propiedad Nombre (Solo lectura en este caso)
    @property
    def nombre(self):
        return self._nombre

    # Propiedad Apellido
    @property
    def apellido(self):
        return self._apellido

    @apellido.setter
    def apellido(self, valor):
        self._apellido = valor

    # Propiedad Cédula con validación
    @property
    def cedula(self):
        return self._cedula

    @cedula.setter
    def cedula(self, valor):
        if valor.isdigit() and len(valor) <= 10:
            self._cedula = valor
        else:
            print("Error: Cédula no válida")
            self._cedula = "0000000000"

    # Propiedad Email
    @property
    def email(self):
        return self._email

    @email.setter
    def email(self, valor):
        self._email = valor

    # Propiedad Semestre con validación 1-9
    @property
    def semestre(self):
        return self._semestre

    @semestre.setter
    def semestre(self, valor):
        if valor >= 1 and valor <= 9:
            self._semestre = valor
        else:
            print(f"Error: El semestre {valor} no es válido (1-9)")

    # Propiedad Materias
    @property
    def materias(self):
        return self._materias

    @materias.setter
    def materias(self, valor):
        self._materias = valor

    def mostrar_datos(self):
        print("--- FICHA DEL ESTUDIANTE ---")
        print(f"Nombre: {self._nombre}")
        print(f"Apellido: {self.apellido}")
        print(f"Cédula: {self.cedula}")
        print(f"Email: {self.email}")
        print(f"Semestre: {self.semestre}")
        print(f"Materias: {self.materias}")

# Ejemplo de uso con todos los datos
materias_actuales = ("Programación", "Base de Datos", "Estadística")
estudiante1 = Estudiante("karen", "Alvarez", "1234567890", "karen.p@universidad.edu", 3, materias_actuales)

estudiante1.mostrar_datos()
```
